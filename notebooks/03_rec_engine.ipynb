Applying Stemming 

```python
from nltk.stem.porter import PorterStemmer

ps=PorterStemmer()

def stem(text):

    y=[]

    for i in text.split():
        y.append(ps.stem(i)) #it takes a word from the entry and reduces it down to it's root word.

    return " ".join(y)
```

```python
movies['tags']=movies['tags'].apply(stem)
```

```python
movies['tags'][0]
```

Count Vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer(max_features=5000,stop_words='english') #storing the vocabulary 

vectors=cv.fit_transform(movies['tags']).toarray() #learns the vocabulary and converts the texts to numbers.
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

Cosine Similarity and the Similarity Matrix

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity=cosine_similarity(vectors) #creating the cosine oriented similarity matrix
```

```python
similarity.shape
```

```python
similarity[0]
```

The actual recommendation engine

```python
def recommend(movie):

    movies_index=movies[movies['title']==movie].index[0] #the index of the movie input 

    distances=similarity[movies_index] #the similarity values with respect to this movie

    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1]) #sorting these values to return the top 5 

    recommendations=[]
    for i in movies_list[1:6]:

        recommendations.append(movies.iloc[i[0]].title) #returning the top 5  similar movies 

    return recommendations


```

```python
recommend("Avatar")
```

```python
recommend("Pulp Fiction")
```
