```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movies=pd.read_csv("../data/tmdb_5000_movies.csv")
credits=pd.read_csv("../data/tmdb_5000_credits.csv")
```

```python
movies.head()
```

```python
credits.head()
```

```python
#have to merge the two datasets 

df=pd.merge(credits,movies,on='title')
df.head()
```

```python
df.info()
```

Keeping only necessary columns


```python
df=df[['movie_id', 'title', 'cast', 'crew', 'genres',
        'keywords', 'overview']]
df.info()
```

Handling Missing Values

```python
df.shape
```

```python
df.isnull().sum()
```

Overview is the only column that has null values and only three. with 4800 entries and only 3 empty columns in overview, we can afford
to drop that column. It's a very important column.


```python
#dropping rows based on overview
df.dropna(subset=['overview'],inplace=True)
```

```python
import ast

def convert(obj):
    L=[]

    for i in ast.literal_eval(obj):
        L.append(i['name'])

    return L;
    
    
```

JSON parsing the genre column, to extract all the genres to which the film belongs and replacing the old dictionary with this new
list containing only genre names. By using the above function conert(). We can also apply it to the keyword column as it stores values in
the  same way as genres used to.

```python
df['genres']=df['genres'].apply(convert)
```

```python
df.head(1)
```

```python
df['keywords']=df['keywords'].apply(convert)
```

```python
df.head(1)
```

We can apply this for cast and crew columns as well but for crew we only need the director's name , so for that we can add
an if condition to the above loop 

```python
import ast

def fetch_direc(obj):
    L=[]

    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            
            L.append(i['name'])

    return L;
    
    
```

```python
df['cast']=df['cast'].apply(convert)
df['crew']=df['crew'].apply(fetch_direc)
```

```python
df.head(1)
```

Need to remove spaces from the names in cast and convert the overview column from plain tex to lists

```python
df['cast']=df['cast'].apply(lambda x:[i.replace(" ","") for i in x])
df['crew']=df['crew'].apply(lambda x:[i.replace(" ","") for i in x])
df['overview']=df['overview'].apply(lambda x:x.split())
```

```python
df.head()
```

This is an important step here, 
we will be creating  a new column called tags that stores the info of columns cast,crew,genre,keywords,overview.
and then convert the entry of each row of that column to string
as it is needed for vectorization

```python
df['tags']=df['overview']+df['keywords']+df['cast']+df['crew']+df['genres']
```

```python
df.head()
```

```python
processed_df=df[['movie_id','title','tags']]
processed_df.head()
```

```python
processed_df['tags']=processed_df['tags'].apply(lambda x:" ".join(x))
```

```python
processed_df['tags']=processed_df['tags'].apply(lambda x:x.lower())
```

```python
processed_df.head()
```

```python
processed_df.to_csv("../data/processed_movies.csv")
```
